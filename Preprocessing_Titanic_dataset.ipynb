# Preprocessing the Titanic Dataset

## import the relevant libraries

```python
import pandas as pd
```

## load the Titanic dataset

```python
raw_csv_data = pd.read_csv('Titanic_raw_data.csv')
```

```python
df = raw_csv_data.copy()
df
```

## dropping non critical columns that can be detected logically

### we will drop the 'name' column because it does not has any predictive power

```python
df = df.drop(['Name'], axis = 1)
df.head()
```

### we will drop the 'id' column because it doesn't have any predictive power and already the rows are ordered from 0 to 890

```python
df = df.drop(['PassengerId'], axis = 1)
df.head()
```

### we will drop the Ticket columns because the fare column will show us the economic level of the passenger and it's enough

```python
df = df.drop(['Ticket'], axis = 1)
df.head()
```

## checking for missing values

```python
df.info()
```

### Cabin, Age and Embarked has missing values, we will fill the NaN in Cabin with unknown instead and then create a Deck column to display just the first letter because we don't need to know the number it's enough to know the cabin|

```python
df['Cabin'].fillna('Unknown', inplace=True)
df['Deck'] = df['Cabin'].str[0]
df = df.drop(['Cabin'], axis = 1)
df
```

### since embarked has just 2 missing values we can fill it with any of it's values this will not affect the analysis

```python
df['Embarked'].fillna('Q', inplace=True)
df.info()
```

### we will replace the NaN values in Age with the mean value

```python
mean_age = df['Age'].mean()
df['Age'].fillna(mean_age, inplace=True)
df.info()
```

### now no missing values are shown, we will just put the 'Survived' column at the end because it will be our target

```python
columns_ordered = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare',
       'Embarked', 'Deck', 'Survived']
df_ordered = df[columns_ordered]
df_ordered.head()
```

## Sex

### we will replace male with 1's and female with 0's is the Sex column

```python
df_ordered['Sex'] = df_ordered['Sex'].replace({'male': 1, 'female': 0})
df_ordered.head()
```

## Age

```python
distinct_ages = df_ordered['Age'].unique()
distinct_ages
```

### we can see that there is fractional values in the age, but it's better to have them as discrete values in term of memory usage and performance

```python
df_ordered['Age'] = df_ordered['Age'].astype(int)
distinct_ages = df_ordered['Age'].unique()
distinct_ages
```

### all ages converted sucessfully

```python
df_ordered.head()
```

## Embarked

```python
Embarked_disticnt_values = df_ordered['Embarked'].unique()
Embarked_disticnt_values
```

```python
Embarked_columns = pd.get_dummies(df_ordered['Embarked'])
Embarked_columns = Embarked_columns.astype(int)
Embarked_columns
```

```python
df_ordered = pd.concat([df_ordered, Embarked_columns], axis = 1)
df_ordered = df_ordered.drop(['Embarked'], axis = 1)

df_ordered.head()
```

### we will replace Embarked letters with teh full port name to avoid repetition when creating dummies for Deck and to be more clear

```python
columns_ordered = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare',
        'Deck', 'Survived', 'Cherbourg', 'Queenstown', 'Southampton']
df_ordered.columns = columns_ordered
df_ordered.head()
```

## Deck

```python
Deck_distinct_values = df_ordered['Deck'].unique()
Deck_distinct_values
```

```python
Deck_columns = pd.get_dummies(df_ordered['Deck'])
Deck_columns = Deck_columns.astype(int)
Deck_columns
```

```python
df_ordered = pd.concat([df_ordered, Deck_columns], axis = 1)
df_ordered = df_ordered.drop(['Deck'], axis = 1)

df_ordered.head()
```

```python
columns_ordered = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare',
         'Cherbourg', 'Queenstown', 'Southampton', 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T', 'U', 'Survived']
df_ordered = df_ordered[columns_ordered]
df_ordered.head()
```
